```python
# Make clusters using K-Means of genres to input to the model

from sklearn.cluster import KMeans

data = movie_ratings[cols]
sse = []

for i in list(range(1, 15)):
    clustering_kmeans = KMeans(n_clusters=i, precompute_distances="auto", n_jobs=-1)
    movie_ratings['clusters'] = clustering_kmeans.fit_predict(data)
    sse.append(clustering_kmeans.inertia_)
```

```python
# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list(range(1, 15)), sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

```python
from sklearn.metrics import silhouette_samples

for i, k in enumerate(list(range(5,8))):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    
    # Run the Kmeans algorithm
    km = KMeans(n_clusters=k, precompute_distances="auto", n_jobs=-1)
    labels = km.fit_predict(data)
    centroids = km.cluster_centers_

    # Get silhouette samples
    silhouette_vals = silhouette_samples(data, labels)

    # Silhouette plot
    y_ticks = []
    y_lower, y_upper = 0, 0
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
        ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
        y_lower += len(cluster_silhouette_vals)

    # Get the average silhouette score and plot it
    avg_score = np.mean(silhouette_vals)
    ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])
    ax1.set_xlabel('Silhouette coefficient values')
    ax1.set_ylabel('Cluster labels')
    ax1.set_title('Silhouette plot for the various clusters', y=1.02);
```

Calculate the Cosine similarity based on users to be used in the calculation for scores of each user for each movie.

```python
#Number of users
print('Filtered Users: {0}'.format(ratings.userId.nunique()))
#Number of Movies
print('Filtered Movies: {0}'.format(movies.movieId.nunique()))
```

```python
#Create a mapping for userIds and movieIds

movieid_map = pd.DataFrame(Rating_avg.movieId.unique()).sort_values(by=0)
movieid_map.columns = ['movieId']
movieid_map['moviemap'] = range(1,len(movieid_map)+1)

mapped_ratings = pd.merge(Rating_avg, movieid_map, how='left', on='movieId')


userid_map = pd.DataFrame(Rating_avg.userId.unique()).sort_values(by=0)
userid_map.columns = ['userId']
userid_map['usermap'] = range(1,len(userid_map)+1)

mapped_ratings = pd.merge(mapped_ratings, userid_map, how='left', on='userId')
```

```python
# Create a matrix of user movie ratings

user_movie_matrix = np.zeros((mapped_ratings.userId.nunique(), mapped_ratings.movieId.nunique()))
for line in mapped_ratings.itertuples():
    user_movie_matrix[line[8]-1, line[7]-1] = line[6]
```

```python
#Calculate Sparsity

sparsity = float(len(user_movie_matrix.nonzero()[0]))
sparsity /= (user_movie_matrix.shape[0] * user_movie_matrix.shape[1])
sparsity *= 100
print('Sparsity: {:4.2f}%'.format(sparsity))
```

```python
# define a function to calculate cosine similarity
def similarity(ratings, kind='user', epsilon=1e-9):
    # epsilon -> small number for handling dived-by-zero errors
    if kind == 'user':
        sim = np.dot(ratings, ratings.T) + epsilon
    elif kind == 'item':
        sim = np.dot(ratings.T, ratings) + epsilon
    norms = np.array([np.sqrt(np.diagonal(sim))])
    return (sim / norms / norms.T)
```

```python
user_similarity = similarity(user_movie_matrix, kind='user')
np.shape(user_similarity)
```

```python
similarity = pd.DataFrame(user_similarity, index=userid_map.userId, columns=userid_map.userId)
similarity.head()
```
